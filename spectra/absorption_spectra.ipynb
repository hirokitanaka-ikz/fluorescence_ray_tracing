```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick
import scienceplots
plt.style.use(["science", "nature"])
```

```python
df_pi = pd.read_csv('YbYLF_pi_300K.txt', delimiter='\t', header=None)
df_sigma = pd.read_csv('YbYLF_sigma_300K.txt', delimiter='\t', header=None)
df_pi.columns = ['wavelength', 'fluorescence', 'em_xs', 'abs_xs']
df_sigma.columns = ['wavelength', 'fluorescence', 'em_xs', 'abs_xs']
```

```python
doping_level = 7.5 # percent
Yb_concentration = doping_level / 100 * 1.4e22
df_pi['abs_coeff'] = df_pi['abs_xs'] * Yb_concentration * 1.15
df_sigma['abs_coeff'] = df_sigma['abs_xs'] * Yb_concentration * 1.15
```

```python
fig, ax = plt.subplots()
ax.plot(df_sigma['wavelength'], df_sigma['abs_coeff'])
ax.plot(df_pi['wavelength'], df_pi['abs_coeff'])
```
